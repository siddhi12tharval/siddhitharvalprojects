```python
#import reqired libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#url of site
url='https://www.flipkart.com/search?q=laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=laptop%7CLaptops&requestId=399caac2-ba9d-4418-9495-a8e65a87eeee'
```

```python
#to get url
response=requests.get(url)
```

```python
response.status_code
```

```python
#parse the url through html parser
soup=BeautifulSoup(response.content,'html.parser')
```

```python
#print soup
soup
```

```python
result=soup.findAll('div',{"class":'_2kHMtA'})
```

```python
#put everything in for loop
product_name=[ ]
product_price=[ ]
product_rating=[ ]
review_count=[ ]
product_details=[ ]


for r in result:
    #name
    try:
        product_name.append(r.find('div',{'class':'_4rR01T'}).get_text( ))
    except:
        product_name.append('n/a')
        
    #price
    try:
        product_price.append(r.find('div',{'class':'_30jeq3 _1_WHN1'}).get_text( ))
    except:
        product_price.append('n/a')
        
    #rating    
    try:
        product_rating.append(r.find('div',{'class':'_3LWZlK'}).get_text( ))
    except:
        product_rating.append('n/a')
        
    #review_count   
    try:
        review_count.append(r.find('span',{'class':'_2_R_DZ'}).get_text( ).strip( ))
    except:
        review_count.append('n/a')
            
        
       #details     
    try:
        product_details.append(r.find('div', {'class':'fMghEO'}).get_text( ).strip( ).replace('\n', ', '))
    except:
        product_details.append('n/a')        
        
        
            
```

```python
print(len(product_name))
print(len(product_price))
print(len(product_rating))
print(len(review_count))
print(len(product_details))
```

```python
#dataframe which includes details of laptop for page 1
product_overview=pd.DataFrame({'name':product_name,'price':product_price,'rating':product_rating,'review_count':review_count,'details':product_details})
product_overview
```

```python
# To collect details of laptop from all pages
product_name=[ ]
product_price=[ ]
product_rating=[ ]
review_count=[ ]
product_details=[ ]

for i in range(1,21):
    
    #url
    url='https://www.flipkart.com/search?q=laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=laptop%7CLaptops&requestId=399caac2-ba9d-4418-9495-a8e65a87eeee&page=' +str(i)
    
    #request
    response=requests.get(url)
    
    #soup object
    soup=BeautifulSoup(response.content,'html.parser')
    
    result
    result=soup.findAll('div',{"class":'_2kHMtA'})
    
    #loop through result
    for r in result:
        
        
        #name
        try:
            product_name.append(r.find('div',{'class':'_4rR01T'}).get_text( ))
        except:
            product_name.append('n/a')
        
        #price
        try:
            product_price.append(r.find('div',{'class':'_30jeq3 _1_WHN1'}).get_text( ))
        except:
            product_price.append('n/a')
        
        #rating    
        try:
            product_rating.append(r.find('div',{'class':'_3LWZlK'}).get_text( ))
        except:
            product_rating.append('n/a')
        
    #review_count   
        try:
            review_count.append(r.find('span',{'class':'_2_R_DZ'}).get_text( ).strip( ))
        except:
            review_count.append('n/a')
            
        
       #details     
        try:
            product_details.append(r.find('div', {'class':'fMghEO'}).get_text( ).strip( ).replace('\n', ', '))
        except:
            product_details.append('n/a')        
        
        
    
```

```python
#Dataframe which includes details of laptops from all pages
product_overview=pd.DataFrame({'name':product_name,'price':product_price,'rating':product_rating,'review_count':review_count,'details':product_details})
```
